### Imports

### Данные

```python
df = pd.read_csv("../DATA/AMES_Final_DF.csv")
```

**Мы будем пытаться спрогнозировать значение колонки SalePrice. Разделим данные на две части - признаки X и целевая переменная y.**

```python
X = df.drop('SalePrice', axis=1)
y = df.SalePrice
```

**С помощью scikit-learn разделим и X, и y на обучающий и тестовый наборы данных. Поскольку далее мы будем использовать поиск по сетке, то выделите под тестовые данные 10% от всех данных. Чтобы получить такое же разбиение данных, как и в нашем блокноте, можем использовать random_state = 101.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=101)
```

**Признаки в этом наборе данных имеют различный масштаб и различные единицы измерения. Для оптимальной работы регрессии, выполним масштабирование признаков X. Обратим внимание на то, какие данные подавать на вход для .fit(), а какие данные на вход для .transform().**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

**Мы будем использовать модель "Elastic Net". Создадим экземпляр модели ElasticNet в scikit-learn, используя параметры по умолчанию.**

```python
from sklearn.linear_model import ElasticNet
```

```python
model = ElasticNet()
```

**Модель Elastic Net работает с двумя основными параметрами - alpha и L1_ratio. Создадим словарь с набором различных значений этих параметров, для поиска по сетке. Мы можем выбрать различные значения параметров.**

```python
param_dict = {'alpha' : (0.5, 1, 5, 15, 20, 50, 100), 
              'l1_ratio' : (.01, .1, .15, .30, .5, .7, .9, 1)}
```

**С помощью scikit-learn создадим объект GridSearchCV и запустите поиск по сетке для нахождения наилучших параметров модели, используя обучающие данные (предварительно смасштабированные).

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=model, param_grid=param_dict, cv=15 , 
             scoring='neg_mean_squared_error', verbose=1)

grid_model.fit(X_train, y_train)
```

**Отобразим наилучшую комбинацию параметров для Вашей модели.**

```python
grid_model.best_params_
```

**Оценим работу модели на тестовом наборе данных (предварительно смасштабированном) в 10%, которые модель ещё не видела.**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
y_pred = grid_model.predict(X_test)
mean_absolute_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```
