## 6. Используя признак charges или imb, проверить, выполняется ли центральная предельная теорема. Использовать различные длины выборок n. Количество выборок = 300. Вывести результат в виде гистограмм. Найти стандартное отклонение и среднее для полученных распределений. Сделать выводы.

```python
# Используем 4 значения длинны выборки - 30, 50, 100, 200
arr_of_samples = [] # Массив содержащий выборки разных размеров по 300 раз каждая 
data = df_1['bmi']
for i in (30,50,100,200):
    samples = [np.random.choice(data, size=i, replace=True) for _ in range(300)] # Значения могут пересекаться 
    arr_of_samples.append(samples)
    
```

```python
len(arr_of_samples[3])
```

```python
len(arr_of_samples[0][0])
```

```python
np.mean(arr_of_samples[0][0]
```

```python
# Найдем стандартное среднее и стандартное отклонение в каждой из выборок


arr_of_means = []  
arr_of_std = [] 
for i in range(len(arr_of_samples)):
    temp_means = [] # Среднее каждой выборки
    temp_std = [] # стандартное отклонение каждой выборки
    for j in range(len(arr_of_samples[i])):
        temp_means.append(np.mean(arr_of_samples[i][j]))
        temp_std.append(np.std(arr_of_samples[i][j]))
    arr_of_means.append(temp_means)
    arr_of_std.append(temp_std)
        
    
        
```

```python
# Построим гистограммы каждой из выборок

#Для средних значений
dict_of_means = {i : arr_of_means[i] for i in range(len(arr_of_means))}
df_of_means = pd.DataFrame(dict_of_means)

#Для стандартного отклонения
dict_of_std = {i : arr_of_std[i] for i in range(len(arr_of_std))}
df_of_std = pd.DataFrame(dict_of_std)
```

```python
df_of_std
```

```python
df_of_means.columns = ['n = 30', 'n = 50','n = 100','n = 200']
df_of_std.columns = ['n = 30', 'n = 50','n = 100','n = 200']
```

## 8. Проверить распределения следующих признаков на нормальность: индекс массы тела, расходы. Сформулировать нулевую и альтернативную гипотезы. Для каждого признака использовать KS-тест и q-q plot. Сделать выводы на основе полученных p-значений.

```python
# Гипотеза H0 - Распределение данных соответствующего признака не отличается от эталонного нормального распределения
# Гипотеза H1 - Распределение отличается от эталонного нормального и не является нормальным

# Перед проведение теста Колмогорова-Смирнова нужно обязательно провести Z-преобразование

print(f'Тест Колмогорова-Смирнова для расходов: {stats.kstest(stats.zscore(df_1.charges), "expon")}')
print(f'Тест Колмогорова-Смирнова для bmi: {stats.kstest(stats.zscore(df_1.bmi), "norm")}')
```

```python
# Построение Q-Q Plot для расходов
stats.probplot(df_1.charges, dist="norm", plot=plt)
plt.title("Q-Q Plot для признака 'charges'")
plt.xlabel("Теоретические квантили")
plt.ylabel("Выборочные квантили")
plt.show()
```

```python
# Построение Q-Q Plot для bmi
stats.probplot(df_1.bmi, dist="norm", plot=plt)
plt.title("Q-Q Plot для признака 'bmi'")
plt.xlabel("Теоретические квантили")
plt.ylabel("Выборочные квантили")
plt.show()
```

<font size = 5> В случае **charges** мы получаем p-value сильно меньшее чем 0.05. Это является основанием для отклонения H0 гипотезы. В случае **bmi** мы получили довольно высокое значение p-value ~0.34, что говорит о том, что у нас нет оснований отклонять H0 гипотезу. О том же самом свидетельствуют и Q-Q plot'ы. Для **charges** имеет место сильное отклонение на всех участках графика, в то время как на **bmi** только на концах.

## 13. Загрузить данные из файла “bmi.csv”. Взять оттуда две выборки. Одна выборка – это индекс массы тела людей c региона northwest, вторая выборка – это индекс массы тела людей с региона southwest. Сравнить средние значения этих выборок, используя t-критерий Стьюдента. Предварительно проверить выборки на нормальность (критерий Шопиро-Уилка) и на гомогенность дисперсии (критерий Бартлетта).

```python
df_3 = pd.read_csv('bmi.csv')
```

```python
northwest = df_3[df_3.region == 'northwest']['bmi']
southwest = df_3[df_3.region == 'southwest']['bmi']
```

```python
df_3
```

```python
# Проверям выборки на нормальность (Критерий Шопиро-Уилка)
print(f'Для northwest : {stats.shapiro(northwest)}')
print(f'Для southwest : {stats.shapiro(southwest)}')
```

```python
# Делаем вывод, что данные имеют нормальное распределение 
```

```python
# Проверям гомогенность дисперсий
print(f'Проверка гомогенности: {stats.bartlett(northwest, southwest)}')
```

```python
# p-value все же превышает 0.05 и у нас нет оснований не считать дисперсии выборок примерно одинаковыми
```

```python
# Сравним выборки по t-Критерию Стьюдента
print(f'Сравнение выборок с использованием t-критерия Стьюдента: {stats.ttest_ind(northwest, southwest)}')
```

<font size = 5> **p-value** меньше 0.05, следовательно мы отвергаем H0 гипотезу о равенстве средних двух выборок

##  15. С помощью критерия Хи-квадрат проверить, являются ли переменные зависимыми.

```python
data = pd.DataFrame({'Женат': [89,17,11,43,22,1],'Гражданский брак': [80,22,20,35,6,4],
'Не состоит в отношениях': [35,44,35,6,8,22]})
data.index = ['Полный рабочий день','Частичная занятость','Временно не работает','На домохозяйстве','На пенсии','Учёба']
```

```python
data
```

```python
chi2, p, dof, expected = stats.chi2_contingency(data);

print(f"Хи-квадрат статистика: {chi2}")
print(f"p-значение: {p}")
print(f"Степень свободы: {dof}")
print(f"Ожидаемые частоты:\n{expected}")
```

<font size = 5> Видим, что p-value сильно меньше 0.05 откуда делаем вывод, что следует отклонить гипотезу H0 о незавимости переменных, То есть да - семейное положение влияет на занятость
